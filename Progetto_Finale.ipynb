````python
# Reimposto l'indice del dataframe
df_btc.reset_index(drop=True, inplace=True)

df_btc.head()
````

Elimino le colonne non necessarie dal dataframe e inizio la conversione delle colonne.

````python
# Seleziono le colonne 'Date' e 'Price'
df_btc = df_btc[['Date', 'Price']]

df_btc.head()
````

````python
# Controllo il tipo di formato delle colonne prima della conversione
df_btc.dtypes
````

````python
# Converto la colonna 'Date' in formato datetime
df_btc['Date'] = pd.to_datetime(df_btc['Date'], format='%m/%d/%Y')

# Verifico il risultato
df_btc.dtypes
````

````python
# Trovo le righe che non possono essere convertite in float
non_float_rows = df_btc[~df_btc['Price'].str.match(r'^-?\d+\.?\d*$')]

# Visualizzo le righe precedentemente trovate
non_float_rows
````

````python
# Rimuovo i punti che le virgole dalla colonna 'Price'
df_btc['Price'] = df_btc['Price'].str.replace('[.,]', '', regex=True)

# Converto la colonna 'Price' al formato float
df_btc['Price'] = df_btc['Price'].astype(float)

# Verifico il risultato
df_btc.dtypes
````

Passo a calcolare le statistiche descrittive per i prezzi del Bitcoin. </p>

````python
# Calcolo la deviazione standard del prezzo di Bitcoin
btc_std = round(df_btc['Price'].std(), 2)

print("Deviazione standard del prezzo di Bitcoin:", btc_std)
````

````python
# Calcolo il valore medio del prezzo di Bitcoin
btc_mean = df_btc['Price'].mean()

# Calcolo la deviazione standard del prezzo di Bitcoin in termini percentuali
btc_std_percentage = round((btc_std / btc_mean) * 100, 2)

print("Deviazione standard del prezzo di Bitcoin in termini percentuali:", btc_std_percentage, "%")
````

````python
# Calcolo i rendimenti giornalieri
df_btc['Daily_Return'] = df_btc['Price'].pct_change() * 100

# Calcolo la varianza dei rendimenti giornalieri
btc_variance = round(df_btc['Daily_Return'].var(), 2)

print("Varianza dei rendimenti giornalieri di Bitcoin:", btc_variance)
````

````python
df_btc.head()
````

````python
# Calcolalo la deviazione standard mobile utilizzando una finestra temporale
window_size_std = 30  # Dimensione della finestra mobile 60
btc_std_rolling = df_btc.set_index('Date')['Price'].rolling(window=window_size_std).std()

# Calcola la deviazione standard in percentuale
btc_mean = df_btc.set_index('Date')['Price'].rolling(window=window_size_std).mean()
btc_std_percentage = (btc_std_rolling / btc_mean) * 100

# Calcolo la varianza mobile dei rendimenti giornalieri utilizzando una finestra temporale
window_size_var = 30  # Dimensione della finestra mobile 30
btc_variance_rolling = df_btc.set_index('Date')['Daily_Return'].rolling(window=window_size_var).var()


# Creo una figura con due subplot affiancati
fig, axs = plt.subplots(2, 1, figsize=(13, 7))

# Grafico della deviazione standard mobile (%)
axs[0].plot(btc_std_percentage.index, btc_std_percentage, label='Deviazione Standard Mobile (30 giorni)', color='#ff9900')
axs[0].set_title('Deviazione Standard del Prezzo di Bitcoin')
axs[0].set_xlabel('Data')
axs[0].set_ylabel('Deviazione Standard (%)')
axs[0].legend()
axs[0].grid(True)

# Grafico della varianza mobile
axs[1].plot(btc_variance_rolling.index, btc_variance_rolling, label='Varianza Mobile (30 giorni)', color='#ff9900')
axs[1].set_title('Varianza dei Rendimenti Giornalieri di Bitcoin')
axs[1].set_xlabel('Data')
axs[1].set_ylabel('Varianza')
axs[1].legend()
axs[1].grid(True)

# Mostro il grafico
plt.tight_layout()
plt.show()
````

<p>Dalla visualizzazione si osserva che sia la deviazione standard sia la varianza, espresse in termini percentuali, sono progressivamente diminuite nel tempo. Questo indica una riduzione della volatilità del Bitcoin, suggerendo che il prezzo del Bitcoin è diventato relativamente più stabile nel corso degli anni. </p>

````python
# Calcolo la percentuale di tempo con rendimento <= 5%
time_less_than_5 = (abs(df_btc['Daily_Return']) <= 5).mean() * 100

# Calcolo la percentuale di tempo con rendimento tra 5% e 10%
time_between_5_and_10 = ((abs(df_btc['Daily_Return']) > 5) & (abs(df_btc['Daily_Return']) <= 10)).mean() * 100

# Calcolo la percentuale di tempo con rendimento >= 10%
time_greater_than_10 = (abs(df_btc['Daily_Return']) >= 10).mean() * 100

# Arrotondo la percentuale a due decimali
time_less_than_5 = round(time_less_than_5, 2)
time_between_5_and_10 = round(time_between_5_and_10, 2)
time_greater_than_10 = round(time_greater_than_10, 2)

# Stampo i risultati
print("Percentuale di tempo con rendimento <= 5%:", time_less_than_5, "%")
print("Percentuale di tempo con rendimento tra 5% e 10%:", time_between_5_and_10, "%")
print("Percentuale di tempo con rendimento >= 10%:", time_greater_than_10, "%")
````

````python
# Imposto la colonna 'Date' come indice del DataFrame per calcolare rendimenti settimanali e mensili
df_btc.set_index('Date', inplace=True)

df_btc.head()
````

````python
# Resample del dataframe per settimana e calcolo del prezzo di chiusura settimanale
weekly_close_price = df_btc.resample('W').last()

# Calcolo del rendimento settimanale in percentuale
weekly_return = weekly_close_price['Price'].pct_change() * 100

# Stampo i risultati
print("Rendimento settimanale di Bitcoin:")
print(weekly_return)
````

````python
# Calcolo la percentuale di tempo con rendimento settimanale <= 5%
weekly_time_less_than_5 = (abs(weekly_return) <= 5).mean() * 100

# Calcolo la percentuale di tempo con rendimento settimanale tra 5% e 10%
weekly_time_between_5_and_10 = ((abs(weekly_return) > 5) & (abs(weekly_return) <= 10)).mean() * 100

# Calcolo la percentuale di tempo con rendimento settimanale >= 10%
weekly_time_greater_than_10 = (abs(weekly_return) >= 10).mean() * 100

# Arrotondo la percentuale a due decimali
weekly_time_less_than_5 = round(weekly_time_less_than_5, 2)
weekly_time_between_5_and_10 = round(weekly_time_between_5_and_10, 2)
weekly_time_greater_than_10 = round(weekly_time_greater_than_10, 2)

# Stampo i risultati
print("Percentuale di tempo con rendimento settimanale <= 5%:", weekly_time_less_than_5, "%")
print("Percentuale di tempo con rendimento settimanale tra 5% e 10%:", weekly_time_between_5_and_10, "%")
print("Percentuale di tempo con rendimento settimanale >= 10%:", weekly_time_greater_than_10, "%")
````

````python
# Resample del dataframe per mese e calcolo del prezzo di chiusura mensile
monthly_close_price = df_btc.resample('M').last()

# Calcolo del rendimento mensile in percentuale
monthly_return = monthly_close_price['Price'].pct_change() * 100

# Stampo i risultati
print("Rendimento mensile di Bitcoin:")
print(monthly_return)
````

````python
# Calcola la percentuale di tempo con rendimento settimanale <= 5%
monthly_time_less_than_5 = (abs(monthly_return) <= 5).mean() * 100

# Calcola la percentuale di tempo con rendimento settimanale tra 5% e 10%
monthly_time_between_5_and_10 = ((abs(monthly_return) > 5) & (abs(monthly_return) <= 10)).mean() * 100

# Calcola la percentuale di tempo con rendimento settimanale >= 10%
monthly_time_greater_than_10 = (abs(monthly_return) >= 10).mean() * 100

# Arrotonda la percentuale a due decimali
monthly_time_less_than_5 = round(monthly_time_less_than_5, 2)
monthly_time_between_5_and_10 = round(monthly_time_between_5_and_10, 2)
monthly_time_greater_than_10 = round(monthly_time_greater_than_10, 2)

# Stampa i risultati
print("Percentuale di tempo con rendimento mensile <= 5%:", monthly_time_less_than_5, "%")
print("Percentuale di tempo con rendimento mensile tra 5% e 10%:", monthly_time_between_5_and_10, "%")
print("Percentuale di tempo con rendimento mensile >= 10%:", monthly_time_greater_than_10, "%")
````

Benché il passato non sia un indicatore infallibile del futuro, individuare pattern o ricorrenze può offrire preziosi dettagli agli stakeholder. Questa parte dell'analisi include:</p>

    <li>Calcolo delle correlazioni tra Bitcoin e l'S&P500.</li>
    <li>Calcolo dei rendimenti medi giornalieri di Bitcoin.</li>
    <li>Identificazione dei rendimenti giornalieri massimi e minimi di Bitcoin.</li>
    <li>Calcolo del rendimento cumulato di Bitcoin.</li>
    <li>Determinazione dell'ampiezza del range di prezzo di Bitcoin.</li>
</ul>


````python
# Visualizzo le colonne del DataFrame
df_combined.columns
````

````python
# Resetto l'indice
df_combined = df_combined.reset_index(drop=False)

# Controllo il risultato
df_combined.head()
````

````python
# Definiscco il preriodo di inizio per ogni halving
start_dates = ['2012-11-27', '2016-07-09', '2020-05-13']
````

````python
correlations = []

for start_date in start_dates:
    # Converto la data di inizio in formato datetime
    start_date = pd.to_datetime(start_date)
    
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = start_date + pd.DateOffset(months=18)
    
    # Estraggo i prezzi per il periodo specificato
    period_data = df_combined[(df_combined['Date'] >= start_date) & (df_combined['Date'] <= end_date)]
    
    # Calcolo la correlazione tra i prezzi di Bitcoin e dell'S&P500
    period_correlation = period_data['Price_btc'].corr(period_data['Price_sp500'])
    
    # Aggiungo la correlazione alla lista
    correlations.append(period_correlation)

# Stampo le correlazioni per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"La correlazione tra Bitcoin e S&P500 nei 18 mesi successivi al {start_date} è: {correlations[i]:.2f}")
````

````python
# Lista per memorizzare i rendimenti medi
average_daily_returns = []

for start_date in start_dates:
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Estraggo i rendimenti giornalieri per il periodo specificato
    period_returns = df_btc.loc[start_date:end_date, 'Daily_Return']
    
    # Calcolo il rendimento medio giornaliero per il periodo
    average_return = period_returns.mean()
    average_daily_returns.append(average_return)

# Stampo i rendimenti medi giornalieri per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Rendimento medio giornaliero nei 18 mesi successivi al {start_date}: {average_daily_returns[i]:.2f}%")
````

````python
# Verifico la presenza di dati nulli o anomali
print(df_btc.isnull().sum())
````

````python
# Liste per memorizzare i rendimenti giornalieri massimi e minimi
max_daily_returns = []
min_daily_returns = []

for start_date in start_dates:
    # Calcolo la data di fine (12 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Verifico che le date siano nel formato corretto
    print(f"Analizzando il periodo: {start_date} - {end_date}")
    
    # Estraggo i rendimenti giornalieri per il periodo specificato
    period_returns = df_btc.loc[(df_btc['Date'] >= start_date) & (df_btc['Date'] <= end_date), 'Daily_Return']
    
    # Verifico i dati estratti
    print(period_returns.describe())
    
    # Calcolo il rendimento giornaliero massimo e minimo per il periodo
    max_return = period_returns.max()
    min_return = period_returns.min()
    
    max_daily_returns.append(max_return)
    min_daily_returns.append(min_return)

# Stampo i rendimenti giornalieri massimi e minimi per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Rendimento giornaliero massimo nei 18 mesi successivi al {start_date}: {max_daily_returns[i]:.2f}%")
    print(f"Rendimento giornaliero minimo nei 18 mesi successivi al {start_date}: {min_daily_returns[i]:.2f}%")
````

````python
# Lista per memorizzare le deviazioni standard
stds = []

for start_date in start_dates:
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Estraggo i rendimenti giornalieri per il periodo specificato
    period_returns = df_btc.loc[(df_btc['Date'] >= start_date) & (df_btc['Date'] <= end_date), 'Daily_Return']
    
    # Calcolo la deviazione standard per il periodo
    std_dev = period_returns.std() * 100
    stds.append(std_dev)

# Stampole deviazioni standard per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Deviazione standard nei 18 mesi successivi al {start_date}: {stds[i]:.2f}%")
````

````python
# Ordino il DataFrame per data
df_btc = df_btc.sort_values(by='Date')

# Calcolo i rendimenti giornalieri
df_btc['Daily_Return'] = df_btc['Price'].pct_change()

# Lista per memorizzare i rendimenti cumulati
cumulative_returns = []

for start_date in start_dates:
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Estraggo i rendimenti giornalieri per il periodo specificato
    period_returns = df_btc.loc[(df_btc['Date'] >= start_date) & (df_btc['Date'] <= end_date), 'Daily_Return']
    
    # Calcolo il rendimento cumulato per il periodo
    cumulative_return = (period_returns + 1).prod() - 1
    cumulative_returns.append(cumulative_return)

# Stampo i rendimenti cumulati per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Rendimento cumulato nei 18 mesi successivi al {start_date}: {cumulative_returns[i] * 100:.2f}%")
````

````python
# Liste per memorizzare i range di prezzo
price_ranges = []

for start_date in start_dates:
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Estraggo i prezzi per il periodo specificato
    period_prices = df_btc.loc[(df_btc['Date'] >= start_date) & (df_btc['Date'] <= end_date), 'Price']
    
    # Calcolo il prezzo massimo e minimo per il periodo
    max_price = period_prices.max()
    min_price = period_prices.min()
    
    # Calcolo il range di prezzo
    price_range = max_price - min_price
    price_ranges.append(price_range)

# Stampo i range di prezzo per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Range di prezzo nei 18 mesi successivi al {start_date}: {price_ranges[i]:.2f}")
````

````python
# Liste per memorizzare i range di prezzo in termini percentuali
price_ranges_percent = []

for start_date in start_dates:
    # Calcolo la data di fine (18 mesi dopo la data di inizio)
    end_date = (pd.to_datetime(start_date) + pd.DateOffset(months=18)).strftime('%Y-%m-%d')
    
    # Estraggo i prezzi per il periodo specificato
    period_prices = df_btc.loc[(df_btc['Date'] >= start_date) & (df_btc['Date'] <= end_date), 'Price']
    
    # Calcolo il prezzo massimo e minimo per il periodo
    max_price = period_prices.max()
    min_price = period_prices.min()
    
    # Calcolo il range di prezzo
    price_range = max_price - min_price
    
    # Calcolo il range di prezzo in termini percentuali
    price_range_percent = (price_range / min_price) * 100
    price_ranges_percent.append(price_range_percent)

# Stampo i range di prezzo in termini percentuali per ciascun periodo
for i, start_date in enumerate(start_dates):
    print(f"Range di prezzo in termini percentuali nei 18 mesi successivi al {start_date}: {price_ranges_percent[i]:.2f}%")
````

### Numero dei portafogli

Il seguente file `.csv` contiene il numero dei portafogli creati sul network Bitcoin dal 03/01/2009 al 23/05/2024 ed è stato estratto dal dataset di Bitcoin di Google BigQuery mediante la seguente query:

```sql
WITH first_seen AS (
  SELECT 
    DATE(block_timestamp) AS date, 
    address 
  FROM 
    bigquery-public-data.crypto_bitcoin.transactions, 
    UNNEST(inputs) AS input, 
    UNNEST(input.addresses) AS address 
  WHERE 
    address IS NOT NULL 
  UNION DISTINCT 
  SELECT 
    DATE(block_timestamp) AS date, 
    address 
  FROM 
    bigquery-public-data.crypto_bitcoin.transactions, 
    UNNEST(outputs) AS output, 
    UNNEST(output.addresses) AS address 
  WHERE 
    address IS NOT NULL
)
SELECT 
  date, 
  COUNT(address) AS new_wallets 
FROM 
  first_seen 
GROUP BY 
  date 
ORDER BY 
  date;
BY
  date
ORDER BY
  date;
